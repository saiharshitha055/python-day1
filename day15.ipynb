```python
'''PYTHON TERNARY OPERATORS:The name of python ternary operators are if-else operators
syntax:varname= exp1 if test condition else exp2
Here,the test condition can be either Relational or Logical expression and whose result can be either True or False.
1.If the Test Condition result is True then the PVM executes exp1 and result placed in LHS varname
2.if the test condition is False then the PVM executes exp2 and places the result in LHS varname.'''
```

```python
'''1.Write a python program which will accept two numerical values and find the biggest among them'''
n1=float(input("enter a first value:"))
n2=float(input("enter a second a value:"))
biggest_value=n1 if n1>n2 else n2
print("({},{})={}".format(n1,n2,biggest_value))
```

```python
'''2.Write a python program which will accept the numerical value that is big or small and either both are equal'''
a=float(input("enter the first value:"))
b=float(input("enter the second value:"))
result="first value is big" if a>b else "both are equal" if a==b else "second value is big"
print("({},{})={}".format(a,b,result))
```

```python
'''write a py program which will accept numerical integer value and finds whether is +ve,-ve or 0'''
num=float(input("enter a number:"))
res="Positive" if (num>0) else "Negative" if (num<0) else "Zero" 
print("{} is {}".format(num,res))
```

```python
'''Write a py program which will accept the age of a person and finds whether is eligible for voting or not'''
n=int(input("enter the age:"))
for i in range(1):
    result="eligible for voting" if (n>=18) else "not-eligible for voting"
    print("age of {} is {}".format(n,result))
```

```python
'''write a py program for numerical or float numbers whether it is even or odd'''
n=float(input("enter a number:"))
result="Even" if (n%2==0) else "odd"
print("{} is {}".format(n,result))
```

```python
'''Write a py program  to  tell whether the given number is divisible by 5 or not'''
n=float(input("enter the number:"))
for i in range(1):
    result="divisible by 5" if (n%5==0) else "not-divisible by 5"
    print("{} is {}".format(n,result))
```

```python
'''write a py program which will accept that a number is -ve or invalid output'''
n=int(input("enter the number:"))
result="negative number" if (n<0) else "Invalid input"
print("{} is {}".format(n,result))
```

```python
'''Write a py program which will accept any word and decide whether it is vowel or not:'''
word=input("enter a word:").lower()
result="vowel" if "a" in word or "e" in word or "i" in word or "o" in word or "u" in word else "consonant"
print("{} is {}".format(word,result))
```

```python
'''Write a py program which will accept whether a given word is palindrome or not'''
n=input("enter a word:")
word="Palindrome" if n==n[::-1] else "not-Palindrome"
print("{} is {}".format(n,word))
```

```python
'''Python short-hand operators: they are also known as augemented assignment operators that contains a combination of mathematical and assignment operators in a single step'''
'''examples'''
a=10
b=12
c=a+b
print(c)
```

```python
a=10
b=12
a=a+b
print(a,b)
```

```python
a=10
b=12
a+=b
print(a,b)
```

```python
a=9
b=8
a-=b
print(a,b)
```

```python
a,b=3,4
a*=b
print(a,b)
```
