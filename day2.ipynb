```python
# map functions without lambda functions
def  square(x):
     return x*x
list(map(square,[1,2,3]))
```

```python
#with lambda expression
list(map(lambda x: x * x, [1,2,3]))
```

```python
#filter function without lambda expressions
def is_odd(x):
    return x%2!=0
list(filter(is_odd,[1,2,3,4,5]))
```

```python
#filter function with lambda expressions
list(filter(lambda x:x%2!=0,[1,2,3]))
```

```python
#reduce function without lambda expressions
from functools import reduce
def multiply(a,b):
    return a*b
result=reduce(multiply,[2,7])
print(result)
```

```python
from functools import reduce
result=reduce(lambda a,b: a+b,[4,8])
print(result)
```
