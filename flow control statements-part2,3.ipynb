'''Write a py program which will accept two numerical values and find biggest among them and check for equality'''

```python
a=float(input("enter first number:"))
b=float(input("enter second number:"))
if a>b:
    print(f"{a} is greater than {b}")
else:
    if b>a:
        print(f"{b} is greater than {a}")
    else:
        if a==b:
            print(f"{a} is equal to {b}")
print(f"Program execution completed")
```

```python
'''Write a py program which will accept three numerical values to find the biggest of all and equality too'''
a=float(input("enter first number:"))
b=float(input("enter second number:"))
c=float(input("enter third number:"))
if (a>b) and (a>c):
    print(f"{a} is greater than {b,c}")
else:
    if (b>a) and (b>c):
        print(f"{b} is greater than {a,c}")
    else:
        if (c>a) and (c>b):
            print(f"{c} is greater than {a,b}")
        else:
            if (a==b) and (b==c) and (c==a):
                print("All the numbers are equal.")
print("End of the program")
```

```python
'''Write a py program that accepts a digit and displays its name'''
d=int(input("enter a number:"))
if d==0:
    print(f"{d} is zero")
elif d==1:
    print(f"{d} is one")
elif d==2:
    print(f"{d} is two")
elif d==3:
    print(f"{d} is three")
elif d==4:
    print(f"{d} is four")
elif d==5:
    print(f"{d} is five")
elif d==6:
    print(f"{d} is six")
elif d==7:
    print(f"{d} is seven")
elif d==8:
    print(f"{d} is eight")
elif d==9:
    print(f"{d} is nine")
elif d<0 and d not in [-1,-2,-3,-4,-5,-6,-7,-8,-9]:
    print(f"{d} -ve  number")
elif d in [1,2,3,4,5,6,7,8,9]:
    print(f"{d} is a +ve number ")
else:
    print(f"this is an if-elif statement")
print("program is executed")
```
