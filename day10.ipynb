```python
                                               #some example programs in python
'''area and perimeter of rectangle'''
l=float(input("enter the value of length of the rectangle:"))
b=float(input("enter the value of breadth of the rectangle:"))
area=l*b
perimeter=2*(l+b)
print("="*50)
print("length={}".format(l))
print("breadth={}".format(b))
print("area={}".format(area))
print("perimeter={}".format(perimeter))
print(">"*50)
```
