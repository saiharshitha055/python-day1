```python
#range
# syntax1: range(stop) 
r=range(6)
print(r,type(r))
for val in r:
    print(val)
```

```python
#syntax 2: range(start,stop)
r=range(2,10)
print(r,type(r))
for val in r:
    print(val)
```

```python
#syntax 3: range(start,stop,step)
r=range(2,10,2)
print(r,type(r))
for val in r:
    print(val)
```

```python
r=range(105,220,10)
print(r,type(r))
for val in r:
    print(val)
```

```python
#pre-defines functions in list
''' 1.append() 
syntax: listobj.append(value)'''
lst=[12,"KVSH"]
print(lst,type(lst),id(lst))
lst.append("python")
print(lst,type(lst),id(lst))
```

```python
'''inseert()
syntax: listobj.insert(index,value)'''

lst=[14,"KVSR",True]
print(lst,type(lst),id(lst))
lst.insert(2,99)
print(lst,type(lst),id(lst))
```

```python
lst1=[24,89,True,"mvr"]
print(lst1,type(lst1),id(lst1))
lst1.insert(-100,"VJA")
print(lst1,type(lst1),id(lst1))
```

```python
'''remove()
syntax: listobj.remove(val)'''
lst=["HYD","VJA","VZA"]
print(lst,type(lst),id(lst))
lst.remove("VJA")
print(lst,type(lst),id(lst))
```

```python
'''pop(index)-- indicates index based removal of the values in the list
syntax: listobj.pop(index)'''
lst=["HYD","VJA","VZA"]
print(lst,type(lst),id(lst))
lst.pop(1)
print(lst,type(lst),id(lst))
```

```python
'''pop()
syntax: listobj.pop()'''
lst=[10,20,30,40,50,60]
print(lst,type(lst),id(lst))
lst.pop()
print(lst,type(lst),id(lst))
```

```python
'''del operator
syntax1: del lstobj[Index]---removes one elementfrom list based indexing'''
lst=[50,40,89]
print(lst,type(lst),id(lst))
del lst[2]
print(lst,id(lst))
```

```python
'''syntax 2: lstobj[begin index:endindex:step]'''
lst=[10,20,30,40,50,60,70,80]
print(lst,type(lst),id(lst))
del lst[2:5:2]
print(lst,id(lst))
```

```python
'''syntax 3: del lstobj--removes all items'''
lst=[10,20,30,40,50,60,70,80]
print(lst,type(lst),id(lst))
del lst[::]
print(lst,id(lst))
```

```python
'''clear()
syntax:lstobj.clear()'''
lst=["KVSH","KVSR",False,7+3j]
print(lst,id(lst))
lst.clear()
print(lst,id(lst))
```

```python
'''index()--obtains the index of the first occurence of specific element
syntax: listobj.index(value)'''
lst=[10,40,90,70,89]
print(lst)
lst.index(40)
lst.index(10)
```

```python
'''enumerate()
syntax: enumerate(iterableobj)'''
lst=['parrot','cat','crow','squirrel','bat']
print(lst,type(lst))
for i,v in enumerate(lst):
    print(i,v)
```

```python
lst=[10,20,30,10,50,10]
print(lst,id(lst))
for i,v in enumerate(lst):
    if(v==10):
        print(i,v)
```

```python
'''copy()
syntax : lstobj2=lstobj.copy()'''
lst=[10,20,30,40,50]
print(lst,type(lst),id(lst))
lst1=lst.copy()
print(lst1,type(lst1),id(lst1))
```

```python
'''count()
syntax:lstobj.count(val)'''
lst=[10,20,30,10,50,10]
print(lst,type(lst),id(lst))
print(lst.count(10))
```

```python
'''reverse()
syntax: lstobj.reverse()'''
lst=[10,20,30,40,50]
print(lst,id(lst))
lst.reverse()
print(lst,id(lst))
```

```python
'''sort()
syntax: lstobj.sort()'''
lst=[50,20,30,10,40]
print(lst,id(lst))
lst.sort()
print(lst,id(lst))
```

```python
'''extend()
syntax: lstobj1.extend(lstobj2)'''
lst1=[10,20,30]
lst2=[40,50,60]
print(lst1,lst2)
lst1.extend(lst2)
print(lst1,lst2)
```
