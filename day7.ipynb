```python
'''an example for set data type using the pre defined functions
set cricket Players={"Rohit","Kohli","Rossum"}
set tennis players={"Travis,"Kinney","Rossum"}'''
# a) find the names of all the players who are playing all the types of games
#for this question we use union()
cp={"Rohit","Kohli","Rossum"}
tp={"Travis","Kinney","Rossum"}
print(cp,type(cp))
print(tp,type(tp))
allcptp=cp.union(tp)
print(allcptp,type(allcptp))
```

```python
#b) find the names of the players who are playing both cricket and tennis 
# we use intersection()
cp={"Rohit","Kohli","Rossum"}
tp={"Kinney","Travis","Rossum"}
print(cp,type(cp))
print(tp,type(tp))
bothcptp=cp.intersection(tp)
print(bothcptp,type(bothcptp))
```

```python
'''c) find the names of the players who are playing only cricket but not tennis
for this we use diffrence()'''
cp={"Rohit","Kohli","Rossum"}
tp={"Kinney","Travis","Rossum"}
print(cp,type(cp))
print(tp,type(tp))
onlycp=cp.difference(tp)
print(onlycp,type(cp))
```

```python
'''d) find the names of the players who are playing only tennis but not cricket
for this also we use difference()'''
cp={"Rohit","Kohli","Rossum"}
tp={"Kinney","Travis","Rossum"}
print(cp,type(cp))
print(tp,type(tp))
onlytp=tp.difference(cp)
print(onlytp,type(tp))
```

```python
'''e) find the names of the players who are exclusively playing cricket and tennis 
for this we use symmetric_difference'''
cp={"Rohit","Kohli","Rossum"}
tp={"Kinney","Travis","Rossum"}
print(cp,type(cp))
print(tp,type(tp))
exclcptp=cp.symmetric_difference(tp)
print(exclcptp,type(exclcptp))
```

```python
'''nested sets or inner sets'''
'''case 1: it is not possible to define one set in another set bcoz the set category data types are unhashable(not possible for indexing and slicing+not possible to modify)'''
s1={14,"KVSh",{13,17,33,77},{45,89,76,55},"PYTHON"}
print(s1,type(s1))
```

```python
'''case2: it is also not possible to define one list in another set as gives the same result as the above'''
s2={90,"KVSR",[13,89,0,76],[33,67,89],99}
print(s2,type(s2))
```

```python
'''case 3: it is possible to define one tuple in another set bcoz tuples are immutable'''
s3={56,"KVSR",(888,9999,777),(34,67,48,4),"NRI"}
print(s3,type(s3))
for val in s3:
    print(val,type(val))
```

```python
'''case4: it is possible to define one list in another set bcoz lists are mutable and can perform indexing'''
lst=[12,67,90,{34,87,68,0},"nyra"]
print(lst,type(lst))
print(lst[3],type(lst[3]))
lst[3].add("python")
print(lst)
```

```python
'''case5: it is possible to define one set in another tuple bcoz tuples are immutable and allows us to locate set objects by using indexing'''
tpi=(23,48,{"mango","grapes","avacado","papaya"},"NRI")
print(tpi,type(tpi),id(tpi))
print(tpi[2].remove("avacado"))
print(tpi[-2].add(100))
print(tpi,type(tpi),id(tpi))
```

```python
'''frozenset() is used to store multiple values either similar type or diffrent type or both the types in a single obj with unique values'''
'''syntax:frozenset(set/list/tpl/str/bytes/bytearray/range)'''
s1={10,20,30,40,10,30,40,70,90,80}
print(s1,type(s1))
fs1=frozenset(s1)
print(fs1,type(fs1))
```

```python
'''pre-defined functions in frozenset
a) copy()'''
fs1=frozenset({100,200,409,777})
print(fs1,type(fs1),id(fs1))
fs2=fs1.copy()
print(fs2,type(fs2),id(fs1))
```

```python
'''isdisjoint()'''
fs1=frozenset({10,20,30,40,50,10,40})
fs2=({100,400,800})
fs3=({20,78,6})
fs1.isdisjoint(fs3)
```

```python
'''issuperset()'''
fs1=frozenset({10,20,30,40,50,70})
fs2=frozenset({10,20,30})
fs1.issuperset(fs2)
```

```python
'''issubset()'''
fs1=({10,20,30,40,50})
fs2=({10,20,30})
fs1.issubset(fs2)
```

```python
'''union()'''
fs1=frozenset({10,20,30,40})
fs2=({100,200,300})
fs1.union(fs2)
```

```python
'''intersection()'''
fs1=({10,20,30,40,50})
fs2=({10,100,200})
fs1.intersection(fs2)
```

```python
'''difference()'''
fs1=frozenset({10,20,30,40,50,60})
fs2=({100,200,300})
fs1.difference(fs2)
```

```python
'''symmetric_difference()'''
fs1=({10,20,30,40,50})
fs2=({100,200,300})
fs3=fs1.symmetric_difference(fs2)
print(fs3,type(fs3))
```

```python
'''nested or frozensets
case 1: it is possible to define one frozenset in another frozenset'''
s1=frozenset({12,"Guido",frozenset({12,78,90}),"VGJC"})
for v in s1:
    print(v,type(v),type(s1))
```

```python
'''case 2: it is not possible to define one list in another frozenset bcoz frozensets are unhashable type'''
s1=frozenset({13,"albert",[14,78,65],[44,88,22],"PPS"})
for v in s1:
    print(v,type(v),type(s1))
```

```python
'''case 3: it is possible to define one tuple in another frozenset bcoz tuples are immutable'''
s1=frozenset({15,"BURNO",(34,56,87,90),(23,48,56,32),"VGJC"})
for v in s1:
    print(v,type(v),type(s1))
```

```python
'''case 4: it is not possible to define one frozenset in another frozenset bcoz sets are unhashable'''
s1=frozenset({11,"harry",{13,65,88},{33,77,55},"NRi"})
print(s1,type(s1))
```

```python
'''case 5: it is possible to define one frozenset in another list bcoz list are mutable and allows us to locate set objects by using indices'''
s1=frozenset[11,"Nina",frozenset({12,78,0}),frozenset({11,12,14}),"NRI"]
for v in s1:
    print(v,type(v),type(s1))
```

```python
'''case 6: it is possible to define one frozenset in another tuple'''
s1=(11,12,frozenset({12,78,90}),frozenset({12,16,18}),"NRI")
for v in s1:
    print(v,type(v),type(s1))
```
