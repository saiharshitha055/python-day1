```python
#types of copy techniques in python
'''shallow copy
the intial content of both the objects are same
the memory address of both the objects are different
the modifications are independent.
syntax: obj2=obj1.copy()'''
lst=[200,"KVSH",56.9,]
print(lst,type(lst),id(lst))
lst1=lst.copy()
print(lst1,type(lst1),id(lst1))
```

```python
lst1.append("ruby")
lst1.insert(2,"myra")
print(lst1,type(lst1),id(lst1))
```

```python
'''deep copy()
the intial content of both the objects are same
the memory address of the both the objects are same
the changes are dependent
syntax: obj1=obj'''
lst=[400,"KVSR",89.09,"java"]
print(lst,type(lst),id(lst))
lst1=lst
print(lst1,type(lst1),id(lst1))
```

```python
'''inner or nested list
the process of defining one list inside another list is called inner or nested list.
syntax: lstobj=[val1,val2,....[val11,val12,...val1n],[val21,val22,...val2n],...val-n]'''
lst=[12,"KVSR",[45,87,65],[90,42,48,75],"RGUKT"]
print(lst)
for val in lst:
    print(val,type(val),type(lst))  
```
