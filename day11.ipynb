```python
'''program for calculating the power of a number'''
b=int(input("enter the base number:"))
e=int(input("enter the degree of the exponent:"))
result=1
for _ in range(e):
    result=result*b
print(f"the result using a loop is:{result}")
print(f"the base number is:{b}")
print(f"the exponent number in the form of degree is:{e}")
```

```python
def power(base,exponent):
    result=1
    for _ in range(exponent):
        result=result*base
    return result
print("the result using a function is:{}".format(result))
```

```python
'''prime number check'''
n=int(input("enter a number:"))
if n>1:
    for i in range(2,int(n**0.5)+1):
        if (n%i)==0:
            print(f"{n} is not  a prime number:")
            break
    else:
        print(f"{n} is  a prime number:")
else:
    print(f"{n} is not a prime number:")
```

```python
'''amstrong number check'''
'''to understand the amstrong number logic we need to follow 3 steps
1.count the digits
2.break the number apart to calculate the power of each digit
3.compare the total to the sum of power of each digit7u'''
n=int(input("enter a number:"))
original=n
sum_val=0
order=len(str(n))
while n>0:
    digit=n%10
    sum_val+=digit**order
    n=n//10
if sum_val==original:
    print(f"{original} is an Amstrong number")
else:
    print(f"{original} is not a Amstrong number")    
```

```python
'''fizz buzz program'''
num1=int(input("enter the first divisor:"))
num2=int(input("enter the second divisor:"))
for i in range(1,21):
    if i%num1==0 and i%num2==0:
        print("fizz buzz")
    elif i%num1==0:
        print("fizz")
    elif i%num2==0:
        print("buzz")
    else:
        print(i)
```

```python
'''program for finding the duplicates of a string in a list'''
list=['h','a','r','s','h','i','t','h','a']
duplicates=set()
for items in list:
    if list.count(items)>1:
        duplicates.add(items)
print("Duplicates:", duplicates)
```
