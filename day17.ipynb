```python
'''Flow Control Statements In Python
The purpose of flow control statements is to perform operations only once or to repeat finite number until the condition becomes false
In this context,we have 3 types of flow control statements
1.Conditional or Selection or Branching Statements
2.Looping or Iteration Statements
3.Transfer Flow Statements'''
```

```python
#1.Conditional Statements
'''The purpose of conditional statements is to perform X operation in the case of True and Y operation in the case of False
We have 4 types of conditional statements in python
a)simple if statement
b)if-else statement
c)if-elif-else statement
d)match case statement'''
```

```python
'''a)simple if statement
Here if is a keyword and the test condition can be either True or False,if the test condition returns True then PVM executes indentation block of satements and later execute other statements or if the test condition returns False then PVM executes without indentation block of statements'''
```

```python
'''Write a python program that accepts two numerical values and decide biggest and check for equality also'''
a=float(input("enter first number:"))#using simple-if statement
b=float(input("enter second number:"))
if (a>b):
    print("Big({},{})={}".format(a,b,a))
if(b>a):
    print("Big({},{})={}".format(a,b,b))
if(a==b):
    print("Both numbers are equal")
```

```python
'''Write a py prog that accepts three numerical values and decide the biggest of all and check the equality too'''
a=float(input("enter first number:"))
b=float(input("enter second number:"))
c=float(input("enter third number:"))
if (a>b) and (a>c):
    print("(Big{},{},{})={}".format(a,b,c,a))
if(b>a) and (b>c):
    print("(Big{},{},{})={}".format(a,b,c,b))
if(c>a) and (c>b):
    print("(Big{},{},{})={}".format(a,b,c,c))
if (a==b) and (a==c)and (c==a):
    print("All numbers are equal")
```
