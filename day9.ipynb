```python
'''print() function is used to display the result of the python program in console
syntax1: print(val1)
         print(val1,val2,....val-n)'''
a=10
b=20
c=a+b
print(a,b,c)
```

```python
'''syntax2: print(msg)
            print(msg1,msg2,...msg-n)
            print(msg1+msg2+....+msg-n)
here the messages are of the string type and this syntax displays the string message on the console            '''
print("Hello Python")
print('Hello Python')
print('''Hello Python''')
print("""Hello Python""")
print("Hello","Python")
print('Hello',"Python")
print("Hello"+"Python")
print("Hello"+" "+"Python")
print("Python"+3.12)
```

```python
'''syntax5: print(messages cum values with format specifiers)
here %d represents int data
%f represents float data
%s represents str data'''
a=10
b=20.9
c=a+b
print("sum of %d and %f is %f" %(a,b,c))
print("sum of %d and %0.2f is %0.3f" %(a,b,c))
```
