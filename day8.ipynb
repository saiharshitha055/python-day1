```python
'''2.pop():
syntax:dict.pop(key): represents the function for removl of (key,value) from non-empty dict object and if the value does not exiat we get key error'''
a2={"name1":"harshi","name2":"rani","name3":"vyshu","name4":"sneha"}
print(a2,type(a2))
a2.pop("name3")
```

```python
'''3.popitem():
syntax:dictobj.popitem() represents the function for removal of last (key,value) from non-empty dict object and if we call this obj on empty dict we get key error'''
a3={10:2,12:4,16:8,20:9,40:8}
print(a3,type(a3),id(a3))
a3.popitem()
```

```python
'''4.copy(): 
syntax:dictobj2=dictobj1.copy() is the function used for copying the content of one dict obj to another dict obj'''
a4={11:1,12:2,13:3,14:4,15:5,16:6}
print(a4,type(a4),id(a4))
a4=a3.copy()
print(a3,type(a3),id(a3))
```

```python
'''5.get()
syntax:dictobj.get(key) this function is used for obtaining value of value by passing the value of key and the value of key does not exist then we get none as result'''
a6={10:1.2,20:2.2,30:3.2,40:4.2}
print(a6,type(a6))
val=a6.get(20)
print(val)
```

```python
'''6.keys() this function is used for obtaining values of key and placed in varname and whose type <class,'dict_keys'>'''
b1={10:"Python",20:"Django",30:"Flask",40:"C++"}
print(b1,type(b1))
keys=b1.keys()
print(keys,type(keys))
for k in keys:
    print(k)
```

```python
'''7.values()
syntax:dictobj.values() this function is used for obtaining values of value and placed in varnameand whose type <class,'dict_values'> '''
b2={10:"Python",20:"Django",30:"Flask",40:"C++"}
print(b2,type(b2))
values=b2.values()
print(values,type(values))
for v in values:
    print(v)
```

```python
'''8.items()
syntax:dictobj.items() this function is used for obtaining (key,value) from dict obj and placed in varname and whose type<class,'dict_items'>'''
b3={10:"Python",20:"Django",30:"Flask",40:"C++"}
print(b3,type(b3))
items=b3.items()
print(items,type(items))
for k,v in items:
    print(k,v)
```

```python
'''nested dict or inner dict
case1: dict in dict is possible to define'''
a1={"sno":48,"sname":"KVSH","Intmarks":{"Maths":80,"Phy":92,"SQL":75},"Extmarks":{"Maths":78,"Phy":88,"SQL":90},"cname":"NRIIT"}
print(a1,type(a1))
```

```python
for its in a1.items():
    print(its)
```

```python
'''case 2: set in dict is also possible'''
a2={"sno":48,"sname":"KVSH","Intmarks":{80,92,75},"Extmarks":{78,88,90},"cname":"NRIIT"}
print(a2,type(a2))
```

```python
for k,v in a2.items():
    print(k,v,type(v),type(a2),id(a2))
```

```python
'''case3: tuple in dict is also possible'''
a3={"sno":48,"sname":"KVSH","Intmarks":(80,92,75),"Extmarks":(78,88,90),"cname":"NRIIT"}
print(a3,type(a3))
```

```python
for k,v in a3.items():
    print(k,"---->",v,"---->",type(v),type(a3),id(a3))
```

```python
'''case4:list in dict is also possible'''
a4={"sno":48,"name":"KVSR","heights":[170,175,180],"weights":[33,77,90],"cname":"NRIIT"}
print(a4,type(a4))
```

```python
for k,v in a4.items():
    print(k,"--->",v,"---->",type(v),type(a4),id(a4))
```

```python
'''case5: dict in set is not possible'''
a5={48,"KVSR",{"P":99.9,"C":55.5,"Java":44.0},"NRIIT"}
```

```python
'''case6: dict in tuple is possible'''
a6=(48,"KVSR",{"P":99.9,"C":55.5,"Java":44.0},"NRIIT")
print(a6,type(a6))
```

```python
for val in a6:
    print(val,type(val),type(a6),id(val))
```

```python
'''case7: dict in list is possible'''
a7=[48,"KVSR",{"P":99.9,"C":55.5,"Java":44.0},"NRIIT"]
print(a7,type(a7))
```

```python
for val in a7:
    print(val,type(val),type(a7),id(a7),id(val))
```
