```python
'''Membership operators: These operators are used to find the existence of a value in a iterable object.
Here,the iterable object means an object that contain more than one value such as sequence ,list,tuple,dict category data types
We have two types of membership opeartors.They are:
1)in operator:This operator is used to find the existence of a value in iterable object.It returns true if the value is found in iterable object otherwise it returns false
syntax:value in iterable-object
2)not in operator:This operator is used to find the non-existence of a value in a iterable object.It returns false if the value is found in an iterable object
syntax:value not in iterable object '''
```

```python
'''Identity Operator:The purpose of this operator is that to compare the memory address of two objects
In python we have two types of identity operators:
1)is operator:This operator returns True if the provided memory address of obj1 and obj2 are same.
syntax:obj1 is obj2
2)is not operator:This operator returns False if the provided memory address of both the objects are same
syntax:obj1 is not obj2'''
```

```python
'''Note:All Deep copy objects with "is" operator returns True
All Deep Copy objects with "is not" operator returns False
All Shallow Copy objects with "is" operator returns False
All Shallow Copy objects with "is not" operator returns True '''
```

```python
list=[12,"KVSR"]
list1=list
print(list,id(list))
print(list1,id(list1))#implements shallow copy
list is list1
```

```python
list1 is list
```

```python
list=[14,"KVSR"]
list1=list.copy()
print(list1,id(list1))#implements deep copy
print(list,id(list))
list1 is list
```

```python
'''NONE data type:'''
a=None
b=None
print(a,id(a))
print(b,id(b))# implements shallow copy
a is b
```

```python
'''b)DICT data type:'''
d1={10:"Apple",20:"Mango",30:"Grapes"}
d2={10:"Apple",20:"Mango",30:"Grapes"}
print(d1,id(d1))
print(d2,id(d2))# implements deep copy
d1 is d2
```

```python
d1 is not d2
```
