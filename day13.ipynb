```python
'''bit-wise operators
the bitwise operators are used to perform operations on interger data in the form of bit by bit.
bitwise operators are applicable on integer values only but not applicable on float point values bcoz floating point values does not contain certainty where as integer data contains the certainty.
the execution process of bit-wise operators:
a)bitwise operators converts integer data into binary bits.
b)bitwise operators applied on binary bits and performs operation depends on type of bitwise operators
c)finally the result of bitwise operators displays the python execution env in the form of decimal number system'''
```

```python
'''there are 6 types of bit-wise operators
<<----bitwise left shift operator
>>----bitwise right-shift operator
&-----bitwise AND operator
|-----bitwise OR operator
~-----bitwise complement operator
^-----bitwise XOR operator'''
```

```python
'''bitwise left shift operator(<<):
syntax:varname=given no<<no.of.bits
The execution process of this operator is that it moves no.of.bits towards left side by adding no.of.bits towards the right side'''
a=10
b=3
c=a<<b
print(c)
```

```python
a,b=8.9,8
c=a<<b
print(c)
```

```python
print(-5<<7)
```

```python
print(66<<-8)
```

```python
'''to calculate the result of the bit-wise left shift operators we do have a formulae:
i.e. result=given data*(2**no.of.bits)'''
```

```python
'''2.bitwise right shift operators(>>): the execution process of these operators are that the no.of bits moves towards the right side by adding no.of zeros to the left side
syntax: varname=given data//2**no.of.bits (but this formula is only applicable for positive values not for the negative values)'''
a=10
b=3
c=a>>b
print(c)
```

```python
a,b=6.5,4
c=a>>b
print(c)
```

```python
a=-19
b=6
c=a>>b
print(c)
```

```python
a=12
b=-8
c=a>>b
print(c)
```

```python
'''* BIT-WISE COMPLEMENT OPERATOR*(~):
syntax: varname=~given number
the execution process of bit-wise complement operator is that it inverts the bits
The formula for bitwise complement operator is ~given number=-(given number+1)'''
```

```python
a=10
print(~a)
```

```python
a=100
print(~a)
```

```python
a=-98
print(~a)
```
